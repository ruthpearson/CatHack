```python
# read dictionaries
file = '../Data/big_cats/processed_data/dict_2visits_case_ARKS_by_cut_off.pkl'
with open(file, "rb") as f:
    dict_2visits_case_ARKS_by_cut_off = pickle.load(f)
    
file = '../Data/big_cats/processed_data/dict_2visits_control_ARKS_by_cut_off.pkl'
with open(file, "rb") as f:
    dict_2visits_control_ARKS_by_cut_off = pickle.load(f)
    
# read cases and controls csv
file = '../Data/big_cats/processed_data/CKD_death_model_data.csv'
model_data = pd.read_csv(file)
```

```python
def filter_data_for_cutoff(cutoff_days, dict_ARKS_by_cutoff, data_in):
    ARKS_list = dict_ARKS_by_cutoff[cutoff_days]
    data = data_in[data_in['ARKS No'].isin(ARKS_list)]
    return data
```

```python
cases = filter_data_for_cutoff(360, dict_2visits_case_ARKS_by_cut_off, model_data)
controls = filter_data_for_cutoff(360, dict_2visits_control_ARKS_by_cut_off, model_data)
data = cases.append(controls)

print(f"num cases: {cases['ARKS No'].nunique()}")
print(f"num controls: {controls['ARKS No'].nunique()}")
```

```python
rng = np.random.default_rng(2022)
```

```python
def by_ARKS_train_test_split(data, test_percent, rng):
    ARKS_list = data['ARKS No'].unique()
    num_test = int(test_percent*len(ARKS_list))
    test = rng.choice(ARKS_list, size=num_test, replace=False)
    train = set(ARKS_list) - set(test)
    return data[data['ARKS No'].isin(train)], data[data['ARKS No'].isin(test)]
```

```python
train_cases, test_cases = by_ARKS_train_test_split(cases, 0.3, rng)
train_controls, test_controls = by_ARKS_train_test_split(controls, 0.3, rng)

train = train_cases.append(train_controls)
test = test_cases.append(test_controls)
```
