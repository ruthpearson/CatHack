####  PLAN
- take cats with more than 3 visits
- for each chunk of 3 visits check last visit:
  -- if cat has CKD put chunk into CKD training pile
  -- take a sample of of non-CKD chunks for training pile too

```python
# fill nans with zero (make sure this happend AFTER the CKD labels are created!)
data.fillna(0, inplace = True)
```

```python
# get list of patients patients with > n unique dates
def getPatientListByMinimumUniqueTestDates(data, originalPatientList, threshold):
    subPatientList = []
    for patient in originalPatientList:
        dataForThisPatient = data.loc[data['Patient Code'] == patient]
        uniqueTestDatesLen = len(dataForThisPatient)
        if uniqueTestDatesLen >= threshold:
            subPatientList.append(patient)
    return subPatientList        

def getSubArraysOf_N(array, n):
    subArrays = []
    i = 0
    j = n
    iterations = len(array)-n+1
    for iterations in range(iterations):
        chunk = array[i:j]
        subArrays.append(chunk)
        i += 1
        j += 1
    return subArrays  

def createDataVectorsNVisits(patients, data, chosenTests, n):
    x_data = []    
    for patient in patients:
        dataForThisPatient = data.loc[data['Patient Code'] == patient]
        testDates = dataForThisPatient['Date'].tolist()
        testDates.sort()
        chunks = getSubArraysOf_N(testDates, n)
        for chunk in chunks:
            x_element = createXelements(chunk, dataForThisPatient, chosenTests, n)
            # add patient hash to x elements
            x_element = [patient] + x_element
            x_data.append(x_element)
    column_names_x = get_column_names(chosenTests, n)
    df_x = pd.DataFrame(x_data, columns=column_names_x)        
    return df_x   

def get_column_names(chosenTests, n):
    # get x_data index lists for 3 visit chunks
    x_data_label = []
    test_set_labels = []
    # make test set labels
    for test in chosenTests:
        test_set_labels.append(test)
        #test_set_labels.append(test + ' test present?')    

        
    if n == 3: 
        x_data_label = x_data_label + [s + ' visit 1' for s in test_set_labels]
        x_data_label = x_data_label + [s + ' visit 2' for s in test_set_labels]
        # add patient code at the start of x
        x_data_label = ['Patient Code'] + x_data_label + ['CKD? visit 3']
        
    elif n == 2: 
        x_data_label = x_data_label + [s + ' visit 1' for s in test_set_labels]
        # add patient code at the start of x
        x_data_label = ['Patient Code'] + x_data_label + ['CKD? visit 2']
    
    
    return x_data_label


def createXelements(chunkVector, patientData, testList, n):
    x = []
    #get the x info from tests
    finalVisitData = patientData.loc[patientData['Date'] == chunkVector[n-1]]
    final_CKD = finalVisitData['CKD?']
    for visit in chunkVector[:-1]: # don't add the data from the last visit
        thisVisitData = patientData.loc[patientData['Date'] == visit]
        for test in testList:
            result = thisVisitData[test]
            x.append(result.values[0])
    # append the final visit CKD value
    x.append(final_CKD.values[0])
    return x
```

```python
chosenTests = list(data.columns)
chosenTests.remove('Patient Code')
chosenTests.remove('Date')
```

```python
n = 2
uniquePatients = data['Patient Code'].unique()
patients_nplus_visits = getPatientListByMinimumUniqueTestDates(data, uniquePatients, n)
```

```python
df_x = createDataVectorsNVisits(patients_nplus_visits, data, chosenTests, n)
```
