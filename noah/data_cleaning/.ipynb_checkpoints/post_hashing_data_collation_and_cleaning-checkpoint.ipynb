### TOTAL DATA SET (USE OUTER)

```python
# inner join on asap and everything else:  only records where blood tests are present
parse_dates = ['Date', 'Date of Birth']
inner = pd.read_csv('../../data/vetAI/de-identified/merged_historical_patients_sdma_asap-inner.csv', parse_dates=parse_dates, infer_datetime_format = False)
inner
```

```python
# outer join on asap and everything else:  joins blood tests onto history records
parse_dates = ['Date', 'Date of Birth']
outer = pd.read_csv('../../data/vetAI/de-identified/merged_historical_patients_sdma_asap-outer.csv', parse_dates=parse_dates, infer_datetime_format = False)
outer
```

### CLEAN UP 

```python
# see how many nans are in column
column = 'Date'
len(outer) - outer[column].count()
```

```python
# see how many 'ID' entries are nan
# this tells us how many of the blood tests didn't match to a record
print len(outer) - outer['ID'].count()
# take only rows with finite 'ID' (remove the 31 nans)
outer.dropna(subset=['ID'], inplace=True) 
```

```python
list(outer.columns)
```

```python
# make a list of columns to remove and remove them
removeColumns = ['age_sex', 'animal_id', 'clinic_ref', 'LatestWeight', 'filename', 'RR', 'HR', 'FM','Unnamed: 0', 'First name', 'Discount', 'Microchip #', 'Species', 'created', 'modified_x', 'Last vaccination', 'Next vaccination', 'time', 'VaccIntMonths', 'VaccIntDays', 'Next vaccination', 'Label 1 data', 'Label 2 data', 'Label 3 data', 'Label 4 data', 'Client code', 'Full name', 'Contact last', 'Surname first', 'Practice name_x', 'Hospital', 'No. found', 'record number', 'Info sheet name', 'OverDue', 'Hold name', 'Service Category', 'patients service list', 'label 1', 'label 2', 'label 3', 'label 4', 'Vaccine', 'Vaccine Name', 'NextVaccine Name', 'vacc reminder', 'refercode', 'correcttext', 'insurer', 'insurancepolicy','insurancenote', 'HoldTxt', 'HoldCode', 'smsopt', 'remindercounter', 'overduereminder', 'modified_y', 'Given Reason', 'Vet', 'History', 'Practice name_y', 'record_no', 'found_count', 'HL7code', 'Pathology', 'Unnamed: 0.1', 'Unnamed: 1', ' ']
totalData = outer.drop(removeColumns, axis=1)
```

```python
list(totalData.columns)
```

```python
# make 'species_breed' column lower case
totalData['species_breed'] = totalData['species_breed'].str.lower()
# remove all canine entries
doges = totalData[totalData['species_breed'].str.contains("canine") == True]
# doges at indexes 19465 and 19474 so drop these
#totalData.drop(totalData.index[[19465, 19474]], inplace = True)
# NOTE: removing nan 'ID' may have removed the pesky dodge snouts already
```

```python
# drop date and species_breed columns
totalData = totalData.drop(['date', 'species_breed'], axis=1)

# drop entries where Date is nan
totalData.dropna(subset=['Date'], inplace = True) 

# drop age columns and add an age at visit column based on date of birth and visit date
totalData = totalData.drop(['Age', 'Age in wmd'], axis=1)

# make an age at visit column
totalData['Date of Birth'] = pd.to_datetime(totalData['Date of Birth'], errors='coerce') 
totalData['Date'] = pd.to_datetime(totalData['Date']) 
totalData['age at visit'] = totalData['Date'] - totalData['Date of Birth']
totalData['age at visit'] = totalData['age at visit'].apply(lambda d: d.days)
totalData[['Date of Birth', 'Date', 'age at visit']].head()
```

### CLEAN UP COLUMN VALUES

```python
# WEIGHT
totalData['Weight'].replace('kg', '', inplace=True, regex=True)
totalData['Weight'].replace('nan', float('NaN'), inplace=True, regex=True)
totalData['Weight'].replace('~', '', inplace=True, regex=True)
totalData['Weight'].replace('\n', '', inplace=True, regex=True)
totalData['Weight'].replace('5.89 in carrier', '3.89', inplace=True, regex=True)
totalData['Weight'].replace('6.78 in cage', '4.78', inplace=True, regex=True)
totalData['Weight'].replace({'\?': ' '}, inplace=True, regex=True)
totalData['Weight'].replace('540g', '0.54', inplace=True, regex=True)
totalData['Weight'].replace(' ', float('NaN'), inplace=True, regex=True)
```

```python
# TEMPERATURE
totalData['temp'].replace('37.7.', '37.7', inplace=True, regex=True)
totalData['temp'].replace('<32', '32', inplace=True, regex=True)
totalData['temp'].replace('N', float('NaN'), inplace=True, regex=True)
totalData['temp'].replace('NT', float('NaN'), inplace=True, regex=True)
totalData['temp'].replace('WNL', float('NaN'), inplace=True, regex=True)
totalData['temp'].replace('-', '', inplace=True, regex=True)
totalData['temp'].replace(' ', '', inplace=True, regex=True)
totalData['temp'].replace('\?', '', inplace=True, regex=True)
# replace values less than 30 C with nan
mask = totalData['temp'] < 30.
column_name = 'temp'
totalData.loc[mask, column_name] = float('NaN')
```

```python
# DENTAL SCORE
totalData['DS'].replace('\?', '', inplace=True, regex=True)
```

```python
# REMOVE WHITE SPACE FROM BLOOD TEST COLUMNS
totalData.rename(columns=lambda x: x.strip(), inplace=True)
```

```python
# BLOOD TESTS
totalData.replace('\xc3\xa5\xc3\x8a', '', inplace=True, regex=True)
totalData.replace('\*', '', inplace=True, regex=True)
totalData.replace(' ', '', inplace=True, regex=True)
totalData.replace('4.9Redandwhitecells', float('NaN'), inplace=True, regex=True)
totalData.replace('<', '', inplace=True, regex=True)
totalData.replace('>', '', inplace=True, regex=True)
```

```python
# change everything to numeric
numericTests = list(totalData.columns)
numericTests.remove('Breed')
numericTests.remove('Colour')
numericTests.remove('Sex')
numericTests.remove('Desexed?')
numericTests.remove('Patient Code')
numericTests.remove('Deceased?')
numericTests.remove('ID')
numericTests.remove('Sam. app')
numericTests.remove('Date')
numericTests.remove('Date of Birth')
for test in numericTests:
    totalData[[test]] = totalData[[test]].apply(pd.to_numeric)
```

### COMBINE UREA AND BUN COLUMNS

```python
totalData['Urea     mmol/L'] = totalData['Urea     mmol/L'].fillna(0)
totalData['BUN      mmol/L'] = totalData['BUN      mmol/L'].fillna(0)  #if the blanks are nan will need this line first
totalData['Urea/BUN mmol/L'] = totalData['Urea     mmol/L'] + totalData['BUN      mmol/L']
# drop the urea and BUN columns, change zeros back to nans
totalData = totalData.drop(['Urea     mmol/L', 'BUN      mmol/L'], axis=1)
totalData['Urea/BUN mmol/L'].replace(0, float('NaN'), inplace=True)
```

### SAVE CLEAN OUTER DATA TO CSV

```python
# write to csv
totalData.to_csv('../../data/vetAI/clinic_clean.csv')
```
