```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter
import math
```

```python
parse_dates = ['Date', 'Date of Birth']
data = pd.read_csv('..Data/clinic_clean.csv', parse_dates=parse_dates, infer_datetime_format = False)
data['Date'] = pd.to_datetime(data['Date'])
```

```python
hist = []
for patient in data['Patient Code'].unique():
    patientData = data[data['Patient Code'] == patient]
    numVisits = len(patientData['Date'].unique())
    hist.append(numVisits)
```

```python
plt.hist(hist, bins = 100)
plt.ylabel('patients')
plt.xlabel('number of visits')
plt.xlim(0,55)
```

```python
# make creatinine plots for all cats
for patient in data['Patient Code'].unique():
    creat = []
    date = []
    patientData = data[data['Patient Code'] == patient]
    visits = patientData['Date'].unique()
    for visit in visits:
        check = patientData[patientData['Date']==visit]['Creat    umol/L'].values[0]
        if math.isnan(check):
            continue
        else:
            creat.append(check)
            date.append(visit)
    # plot this patient's time series
    plt.plot(date, creat)
plt.hlines(160, pd.Timestamp('2013-10-01'), pd.Timestamp('2018-11-01'), colors='k', linestyles='dashed', zorder=10)
plt.hlines(70, pd.Timestamp('2013-10-01'), pd.Timestamp('2018-11-01'), colors='k', linestyles='dashed', zorder=10)
plt.title('blood creatinine')
plt.ylabel('umol/L')
plt.show()
```

```python
# make urea plots for all cats
for patient in data['Patient Code'].unique():
    creat = []
    date = []
    patientData = data[data['Patient Code'] == patient]
    visits = patientData['Date'].unique()
    for visit in visits:
        check = patientData[patientData['Date']==visit]['Urea/BUN mmol/L'].values[0]
        if math.isnan(check):
            continue
        else:
            creat.append(check)
            date.append(visit)
    # plot this patient's time series
    plt.plot(date, creat)

plt.hlines(10.7, pd.Timestamp('2013-10-01'), pd.Timestamp('2018-11-01'), colors='k', linestyles='dashed', zorder=10)
plt.hlines(5.4, pd.Timestamp('2013-10-01'), pd.Timestamp('2018-11-01'), colors='k', linestyles='dashed', zorder=10)
plt.title('blood urea')
plt.ylabel('mmol/L')
plt.show()
```

```python
# make phosphate plots for all cats
for patient in data['Patient Code'].unique():
    creat = []
    date = []
    patientData = data[data['Patient Code'] == patient]
    visits = patientData['Date'].unique()
    for visit in visits:
        check = patientData[patientData['Date']==visit]['Phos     mmol/L'].values[0]
        if math.isnan(check):
            continue
        else:
            creat.append(check)
            date.append(visit)
    # plot this patient's time series
    plt.plot(date, creat)

plt.hlines(2.26, pd.Timestamp('2013-10-01'), pd.Timestamp('2018-11-01'), colors='k', linestyles='dashed', zorder=10)
plt.hlines(0.9, pd.Timestamp('2013-10-01'), pd.Timestamp('2018-11-01'), colors='k', linestyles='dashed', zorder=10)
plt.title('blood phosphate')
plt.ylabel('mmol/L')
plt.show()
```
